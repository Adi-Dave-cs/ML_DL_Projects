```python
from gensim.models import Word2Vec

# Example corpus — a list of tokenized sentences
sentences = [
    ["the", "cat", "sits", "on", "the", "mat"],
    ["the", "dog", "barks", "at", "the", "cat"],
    ["the", "cat", "meows"],
    ["the", "dog", "runs"],
    ["dogs", "and", "cats", "are", "animals"]
]

# Create and train the Word2Vec model
model = Word2Vec(sentences, vector_size=50, window=2, min_count=1, workers=2, sg=1)

# Show similar words
print("\nMost similar to 'cat':")
print(model.wv.most_similar('cat'))

print("\nMost similar to 'dog':")
print(model.wv.most_similar('dog'))

# See the word vector for 'cat'
print("\nVector for 'cat':")
print(model.wv['cat'])

# Save and load the model to check
model.save("my_word2vec.model")
loaded_model = Word2Vec.load("my_word2vec.model")

print("\nLoaded model check, similar to 'cat':")
print(loaded_model.wv.most_similar('cat'))
```
